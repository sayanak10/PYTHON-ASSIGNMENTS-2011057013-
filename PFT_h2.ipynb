```python
# Develop Arithmetic calculator based on user choice the program should perform arithmetic operation (Any 7 opeations)
# Program make a simple calculator

print("Select operation.")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")
print("5.Exponent")
print("6.Floor Division")
print("7.Modulus")

while True:
    # take input from the user
    choice = input("Enter choice(1,2,3,4,5,6,7): ")

    # check if choice is one of the four options
    if choice in ('1', '2', '3', '4','5','6','7'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(num1+num2)

        elif choice == '2':
            print(num1-num2)

        elif choice == '3':
            print(num1*num3)

        elif choice == '4':
            print(num1/num2)
        elif choice == '5':
          print(num1**num2)
        elif choice == "6":
          print(num1//num2)
        elif choice=='7':
          print(num1%num2)
        
        
        next_calculation = input("Do you want to calculate next (yes/no): ")
        if next_calculation == "no":
          break
        elif next_calculation == "yes":
          continue
        else:
          print("Invalid Input")
          break
    
    else:
        print("Invalid Input")
        break
```

```python
# Develop String Calculator based on user choice the program should perform string operations (Any 5 operations )

print("Select operation.")
print("1.length")
print("2.replace")
print("3.reverse")
print("4.join")
print("5.compare")

while True:
    # take input from the user
    choice = input("Enter choice(1,2,3,4,5): ")

    # check if choice is one of the four options
    if choice in ('1', '2', '3', '4','5'):

        if choice == '1':
            str1=input("Enter the string to find the length")
            print(len(str1))

        elif choice == '2':
            str2=input("Enter a sample string to perform replace operation")
            a=input("Enter the letter which you want to replace")
            b=input("Enter the letter with which you want to replace the previous word")
            print(str2.replace(a,b))

        elif choice == '3':
          str3=input("Enter the sample string to perform the reverse function")
          print(''.join(reversed(str3)))

        elif choice == '4':
          str4 =input("Enter first string")
          str5= input("Enter the second string")
          print(str4.join(str5))
        
        elif choice == '5':
          s1=input("enter the firts string")
          s2=input("enter the second string")
          if (s1==s2):
            print("Both strings are equal")
          else:
            print("both string are not equal")

        next_calculation = input("Do you want to continue (yes/no): ")
        if next_calculation == "no":
          break
        elif next_calculation == "yes":
          continue
        else:
          print("Invalid Input")
          break
    
    else:
        print("Invalid Input")
        break




```
